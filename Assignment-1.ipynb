# Imported libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read Dataset

```python
df1 = pd.read_excel("sales_data_sample.xlsx")
df1.head()
```

```python
df2 = pd.read_json("csvjson.json")
df2.head()
```

```python
df = pd.read_csv("sales_data_sample.csv")
df.head()
```

```python
all_data = pd.concat([df,df1,df2], ignore_index=True)
all_data.head()
```

# Handling Missing Values

```python
all_data['ADDRESSLINE2'].fillna(all_data['ADDRESSLINE2'].mode()[0], inplace=True)
all_data['STATE'].fillna(all_data['STATE'].mode()[0], inplace=True)
all_data['POSTALCODE'].fillna(all_data['POSTALCODE'].mode()[0], inplace=True)
all_data['TERRITORY'].fillna(all_data['TERRITORY'].mode()[0], inplace=True)
```

```python
all_data.isnull().sum()
```

# Removing Duplicates

```python
all_data.duplicated().sum()
```

```python
sales_data = all_data.drop_duplicates()
sales_data.head()
```

```python
sales_data.shape
```

# Data Transformation

```python
sales_data['SALES'] = sales_data['QUANTITYORDERED'] * sales_data['PRICEEACH']
```

```python
total_sales = sales_data['SALES'].sum()
avg_order_value = sales_data['SALES'].mean()
category_distribution = sales_data['PRODUCTLINE'].value_counts()
```

# Data Visualization

# 1. Bar plot

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=category_distribution.index, y=category_distribution.values)
plt.xticks(rotation=45)
plt.xlabel('Product Line')
plt.ylabel('Number of Sales')
plt.title('Product Line Distribution')
plt.show()
```

# 2.Box Plot

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='PRODUCTLINE', y='SALES', data=all_data)
plt.xticks(rotation=45)
plt.xlabel('Product Line')
plt.ylabel('Sales')
plt.title('Sales Distribution by Product Line')
plt.show()
```

# 3.Scatter Plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=sales_data, x='QUANTITYORDERED', y='SALES')
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
plt.title('Scatter Plot: Quantity Ordered vs. Sales')
plt.show()
```

# 4.Pie chart

```python
plt.figure(figsize=(8, 8))
plt.pie(category_distribution, labels=category_distribution.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Sales by Product Line')
plt.show()
```

# 5.Histogram

```python
plt.figure(figsize=(10, 6))
plt.hist(category_distribution, bins=10, edgecolor='black')
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.title('Distribution of Total Sales by Product Line')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# 6.Pair Plot

```python
numeric_columns = ['QUANTITYORDERED', 'PRICEEACH', 'SALES', 'MSRP']
sns.set(style="ticks")
sns.pairplot(sales_data[numeric_columns])
plt.suptitle('Pair Plot of Numeric Variables', y=1.02)
plt.show()
```
